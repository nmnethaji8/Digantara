```python
from calculating_LLA import *
import psutil
import time

## Time Interval Specifications

time_period = date_time(2019, 12, 9, 12, 0, 0)  #start time jday(2019, 12, 9, 12, 0, 0)
days = 5                                        #5 days
interval = 1                                    #1 minute
time_steps = 5*24*60                            #5*24*60 #Total timesteps

# Example usage:
bottom_left = (-60.0, -120.0)                   # Replace with actual coordinates
top_right = (60.0, 120.0)                       # Replace with actual coordinates
```

```python
cpu_utilization = np.zeros(16,float)
time_taken = np.zeros(16,float)
ram_utilization = np.zeros(16,float) 
```

```python

for i in range(16):
    # Reading the satellite names
    satsnames = open('30sats.txt', 'r').readlines()
    num_of_sats = int(len(satsnames)/2)

    cpu_start = psutil.cpu_percent()
    time_start = time.time()
    ram_start = psutil.virtual_memory().percent
    ray.init(num_cpus=i+1)

    LLA = ray.get([single_LLA.remote(satsnames[2*i], satsnames[2*i+1], time_steps, time_period) for i in range(num_of_sats)])
    satellite_times = ray.get([find_entry_exit_times_single.remote(trajectory, bottom_left, top_right) for trajectory in LLA])    

    # Get the cluster resources
    resources = ray.cluster_resources()

    # Get the number of workers
    num_workers = resources.get("CPU", 0)  # You can use a resource name other than "CPU" if needed
    
    print(num_workers)
    cpu_end = psutil.cpu_percent()
    ram_end = psutil.virtual_memory().percent
    
    ray.shutdown()
    time_end = time.time()

    cpu_utilization[i] = cpu_end - cpu_start
    time_taken[i] = time_end-time_start
    ram_utilization[i] = ram_end - ram_start

    # Saving the entry_exit_times in a file
    # saving_entry_exit_times(satellite_times, time_period, bottom_left, top_right)
```

```python
p1 = plt.plot(range(1, 17),time_taken)
p1 = plt.xlabel("Number of workers")
p1 = plt.ylabel("Time taken in seconds")
p1 = plt.title("Time taken vs Workers")
p1 = plt.grid()
```

```python
p2 = plt.plot(range(1, 17),cpu_utilization, label='Percentage of CPU Utilized')
p2 = plt.plot(range(1, 17),ram_utilization, label='Percentage of RAM Utilized')
p2 = plt.xlabel("Number of workers")
p2 = plt.ylabel("Percentage Utilization")
p2 = plt.title("Percentage Utilization vs Workers")
p2 = plt.legend()
p2 = plt.grid()
```
